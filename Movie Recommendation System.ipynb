```python
#Importing required python packages
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer as CntVect
from sklearn.metrics.pairwise import cosine_similarity as CosSim
```

```python
# Read CSV file
movies=pd.read_csv('C:/Users/amrut/Downloads/netflix_titles.csv')
```

```python
movies.head()
```

```python
# Data preprocessing
movies=movies.dropna().reset_index(drop=True)
movies=movies.rename(columns={'listed_in':'genre'})
movies.head()
```

```python
# Feature Selection and Combining them
features=movies[['director','cast','genre']]
features_combined=features['director']+" "+features['cast']+" "+features['genre']
```

```python
features_combined.head()
```

```python
# Count matrix
matrix_count=CntVect().fit_transform(features_combined)
matrix_count=matrix_count.toarray()
```

```python
# Cosine Similarity
similarity=CosSim(matrix_count)
```

```python
#Finding index of movie watched by the user
User_pref="Good People"
movie_index=movies.index[movies['title']==User_pref].values[0]
#movie_index=movies.iloc[movie_index]
movie_index
```

```python
#Recommending movies based on movie watched by the user
recommendations=list(enumerate(similarity[movie_index]))
```

```python
#Sorting in descending order to find movies most related
recommendations=sorted(recommendations,key=lambda x:x[1], reverse=True)
```

```python
#Printing top 10 recommended movies
i=0
for movie in recommendations:
    print(movies['title'].iloc[movie[0]])
    i=i+1
    if(i>10):
        break
```
